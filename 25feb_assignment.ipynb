In pandas, the function '.loc' and 'i.loc' are used to access and subset data in a DataFrame, but they have different ways of indexing and selecting data:


1.) '.loc': 

.'.loc' is label-based indexing, which means it user the labels(row and column names) to access data.

. It takes two arguments separated by comma: the row label(s) and the column label(s).

.The row and column labels can be single labels, lists, or slices.

.The reult of using '.loc' includes both the start and end elements of the slice, if applicable.

```python
df.loc[2,'course_name']
```

```python
df.loc[2,'duration']
```

2.) 'iloc':

. '.iloc' is integer-based indexing, which means it uses the integer positions(row and column indices) to access data.
. It takes two arguments separated by comma: the row index(es) and the column index(es).

.The row and column indices can be single integers, lists, or slices.

.The result of using '.ilocs' includes the start index but excludes the end index,similar to python's slicing



```python
df.iloc[1]
```

#Ans.6)

In the context of pandas, the them "window function" refers to a group of functions that operate on a set of rows(or a "window") of a DataFrame or Series. These function perform calculations or transformation on the data within the window, which is typically defined by a specified range of rows.

Pandas provides various window function that you can use to perform computations on sliding or rolling window of data. Here are some commonly used window function in pandas: 

1.) Rolling Function: These function operate on a rolling window defined by a specified number of preceding  or succeeding rows. Some examples include: 

.rollig(): Computes a rolling mean, sum, min, max, etc.
.rolling().apply(): Applies a custom function to a rolling window.

2.) Expanding FUnctions: These function perform calculations on an expanding window that includes all preceding rows.

.expanding(): Computes an expanding mean, sum, min, max, etc.
.expanding().apply(): Applies a custom function to an expanding window.

3.) Exponentially Weighted Function: These  function compute exponentially weighted statistics based on the values in the window. example

.ewm(): Computes exponentially weifghted mean, sum, variance etc.
.ewm().apply():Applies a custom function to the exponentially weighted window.

4.) GroupBy with Window Function: you can also combine window functions with the 'groupby()' operation to perform calculations on grouped data.

.groupby().rolling():Computes rolling calculations within each group.
.groupby().expanding():Computes expanding calculations within each group.
